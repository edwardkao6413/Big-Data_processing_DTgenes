```python
def create_dataset_RNA(data_gene, data_chr, gene_ver, journal, imf, article, ref, meth, trait, data_value_p = '', data_value_fc = '', data_value_lod = '', rehy_time = '', dt_time = '', omics_type = 'Transcriptomics', growth_stage = '', notes = '', random = ''):
    wgh = pd.DataFrame()
    wgh['methodology'] = meth
    wgh['trait'] = trait
    wgh['population'] = ''
    wgh['marker'] = ''
    wgh['mapped_gene'] = data_gene
    wgh['chr'] = data_chr
    wgh['bp0'] = ''
    wgh['bp1'] = ''
    wgh['p'] = data_value_p
    wgh['lod'] = data_value_lod
    wgh['fc'] = data_value_fc
    wgh['score'] = ''
    wgh['n'] = ''
    wgh['rep'] = ''
    wgh['lab/field'] = random
    wgh['rehydration_time'] = rehy_time
    wgh['drought_treatment_span'] = dt_time
    wgh['對照'] = ''
    wgh['ref'] = ref
    wgh['gene_version'] = gene_ver
    wgh['journal'] = journal
    wgh['if(max)'] = imf
    wgh['omics_type'] = omics_type
    wgh['omics_flag'] = 'Y'
    wgh['analysis_flag'] = 'N'
    wgh['layer'] = 'RNA'
    wgh['growth_stage'] = growth_stage
    wgh['notes'] = notes
    wgh['article_name'] = article
    return wgh
```
